```python
# Project 4: Secure Password Vault 
# Build a Command Line Password Vault that allows users to: 
# 1. Create new credentials for websites. 
# 2. View all stored credentials. 
# 3. Search credentials for a specific website. 
# 4. Validate passwords for strength using for loop and conditions. 
# Features to Include: 
# 1. Menu Options (use if-elif-else): 
# 1. Add New Credential 
# 2. View All Credentials 
# 3. Search by Website 
# 4. Exit 
# 2. Add Credential Flow: 
# o Input: Website name, Username, Password 
# o Use a for loop to validate password: 
# ▪ At least 8 characters 
# ▪ At least 1 uppercase, 1 lowercase, 1 digit, 1 special character 
# credentials = [ 
# {"website": "gmail", "username": "abc@gmail.com", "password": "Strong@123"} 
# ] 
# 3. View Credentials: 
# o Use a for loop to print all saved credentials. 
# 4. Search Credential: 
# o Input: Website name 
# o Use a for loop and if to find and print matching entry. 
# Sample Output: 
# Welcome to Secure Password Vault! 
# 1. Add New Credential 
# 2. View All Credentials 
# 3. Search by Website 
# 4. Exit 
# Enter your choice: 1 
# Enter website name: Instagram 
# Enter username: chetana_insta 
# Enter password: Insta@2024 
# Password Strength: Strong 
# Credential Saved Successfully! 
# Concepts Used: 
# • for loop 
# • if-elif-else 
# • len(), isupper(), isdigit(), islower() 
# • list, dict, input() 
# • String and numeric data types 
```

```python
print("\n -------------------- Welcome to Secure Password Vault ------------------------")

credentials = []   # List to store dictionaries

while True:
    print("\n1. Add New Credential")
    print("2. View All Credentials")
    print("3. Search by Website")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        website = input("Enter Website Name: ")
        username = input("Enter Username: ")
        password = input("Enter Password: ")

        # Password validation
        has_upper = False
        has_lower = False
        has_digit = False
        has_special = False

        if len(password) >= 8:
            for ch in password:
                if ch.isupper():
                    has_upper = True
                elif ch.islower():
                    has_lower = True
                elif ch.isdigit():
                    has_digit = True
                else:
                    has_special = True

        if has_upper and has_lower and has_digit and has_special:
            print("Password Strength: Strong")
        else:
            print("Password Strength: Weak — but saving anyway")

        # Store credentials
        credentials.append({
            "website": website.lower(),
            "username": username,
            "password": password
        })

        print("Credential Saved Successfully!")

    elif choice == 2:
        if not credentials:
            print("No Credentials Saved Yet!")
        else:
            print("\n------ Stored Credentials ------")
            for cred in credentials:
                print(f"Website: {cred['website']}, Username: {cred['username']}, Password: {cred['password']}")

    elif choice == 3:
        search_site = input("Enter Website to Search: ").lower()
        found = False

        for cred in credentials:
            if cred['website'] == search_site:
                print("\nCredential Found:")
                print(f"Website: {cred['website']}")
                print(f"Username: {cred['username']}")
                print(f"Password: {cred['password']}")
                found = True
                break

        if not found:
            print("No Credential Found for this Website.")

    elif choice == 4:
        print("Thank you for using Password Vault!")
        break

    else:
        print("Invalid Choice! Please try again.")
```
