```python
# Project 2:Movie Ticket Pricing System 
# Description: 
# Create a system to calculate movie ticket price based on age and time of show. 
# Rules: 
# • Age < 5: Free 
# • Age 5–12: ₹100 
# • Age 13–60: ₹200 
# • Age > 60: ₹120 
# • If showtime is before 12 PM, apply 20% discount on price 
```

```python
print("\n------- Enter the Details -----------")
age = int(input("Enter the Age: "))
showtime = input("Enter the showtime (in 24-hour format, e.g., 10, 15): ").strip()

price = 0

if age < 5:
    price = 0
    print("Congrats..! Ticket is Free. Enjoy.")
elif 5 <= age <= 12:
    price = 100
elif 13 <= age <= 60:
    price = 200
else:
    price = 120

showtime = int(showtime)

if showtime < 12 and price > 0:
    price -= price * 0.20

print("\n--------- Movie Ticket ------------")
print("Ticket Price is : ₹", price)
```
