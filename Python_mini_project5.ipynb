```python
# Project 5: Quiz Game 
# Objective: 
# Create a command-line Quiz Game that asks multiple-choice questions, takes the user's answers, 
# evaluates the score, and gives a performance review at the end. 
# Features: 
# 1. Show multiple questions with 4 options each. 
# 2. Accept user input (A/B/C/D) and check if itâ€™s correct. 
# 3. Keep track of the total score using a for loop. 
# 4. Show final score and a custom message based on performance. 
```

```python
print("\n------------------- Welcome to the Quiz Game -------------------")

questions = [
    {
        "question": "1. Who developed Python?",
        "options": ["A. Guido van Rossum", "B. Dennis Ritchie", "C. James Gosling", "D. Bjarne Stroustrup"],
        "answer": "A"
    },
    {
        "question": "2. What is the extension of Python files?",
        "options": ["A. .java", "B. .py", "C. .cpp", "D. .txt"],
        "answer": "B"
    },
    {
        "question": "3. Which data type is immutable?",
        "options": ["A. List", "B. Dictionary", "C. Tuple", "D. Set"],
        "answer": "C"
    },
    {
        "question": "4. Which function is used to get input from user?",
        "options": ["A. input()", "B. scan()", "C. read()", "D. scanner()"],
        "answer": "A"
    },
    {
        "question": "5. What is the output of len('Hello')?",
        "options": ["A. 3", "B. 4", "C. 5", "D. 6"],
        "answer": "C"
    },
    {
        "question": "6. Which operator is used for exponent in Python?",
        "options": ["A. ^", "B. **", "C. //", "D. *"],
        "answer": "B"
    },
    {
        "question": "7. What does PIP stand for?",
        "options": ["A. Program Install Package", "B. Python Internal Package", "C. Preferable Package Manager", "D. Pip Installs Packages"],
        "answer": "D"
    },
    {
        "question": "8. Which loop runs at least once?",
        "options": ["A. for loop", "B. while loop", "C. do-while loop", "D. none"],
        "answer": "C"
    },
    {
        "question": "9. Which keyword is used to define a function?",
        "options": ["A. func", "B. method", "C. define", "D. def"],
        "answer": "D"
    },
    {
        "question": "10. Which data structure stores key-value pairs?",
        "options": ["A. List", "B. Tuple", "C. Dictionary", "D. Set"],
        "answer": "C"
    }
]

score = 0

# Quiz execution
for q in questions:
    print("\n" + q["question"])
    
    for opt in q["options"]:
        print(opt)

    user_ans = input("Enter your answer (A/B/C/D): ").upper()

    if user_ans == q["answer"]:
        print(" Correct!")
        score += 1
    else:
        print(" Wrong! Correct answer is", q["answer"])

# Final results
print("\n------------------- Quiz Completed -------------------")
print(f"Your Final Score: {score}/{len(questions)}")

# Performance review
if score == len(questions):
    print(" Excellent! Perfect Score!")
elif score >= 8:
    print(" Great Job! Very Strong Performance!")
elif score >= 5:
    print(" Good! Keep Learning and Improving!")
else:
    print(" Needs Improvement! Practice More.")

print("\nThank you for playing! ")
```
