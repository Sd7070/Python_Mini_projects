```python
# Project 3:Online Food Ordering System 
# Description: 
# Create a Command Line Interface based food ordering system.(script mode) 
# Menu: 
# 1. Pizza - â‚¹150 
# 2. Burger - â‚¹100 
# 3. Fries - â‚¹60 
# 4. Cold Drink - â‚¹40 
# â€¢ Ask user to select items (single choice or multiple) 
# â€¢ Ask quantity for each selected item 
# â€¢ Calculate total bill 
# â€¢ If total bill > â‚¹300, apply â‚¹50 discount 
# â€¢ Display final bill with item details and discount applied (if any) 
# e.g. python food_order.py 
# output: 
# Welcome to Online Food Ordering! 
# Menu: 
# 1. Pizza - â‚¹150 
# 2. Burger - â‚¹100 
# Enter your choice: 1 
# Enter quantity: 2 
# Your total bill is â‚¹300
```

```python
print("\n--------------- Welcome to Online Food Store -------------------------")
print("\n---------------- ***** Menu **** -------------------------------------")
print(" 1. Pizza - â‚¹150 \n 2. Burger - â‚¹100 \n 3. Fries - â‚¹60 \n 4. Cold Drink - â‚¹40")

total = 0
orders = []

while True:
    try:
        choice = int(input("Select item from Menu (1-4): "))
    except ValueError:
        print("Invalid input! Please enter a number.")
        continue

    qty = int(input("Enter Quantity: "))

    match choice:
        case 1:
            item = "Pizza"
            price = 150
        case 2:
            item = "Burger"
            price = 100
        case 3:
            item = "Fries"
            price = 60
        case 4:
            item = "Cold Drink"
            price = 40
        case _:
            print("Invalid choice! Please select item between 1-4.")
            continue

    subtotal = price * qty
    total += subtotal
    orders.append((item, qty, price, subtotal))

    ch = input("Do you want to add more items? (yes/no): ").lower()
    if ch != "yes":
        break

# Discount
discount = 50 if total > 300 else 0
final_amount = total - discount

print("\n-------------- Final Bill -------------------")
for item, qty, price, subtotal in orders:
    print(f"{item} x {qty} = â‚¹{subtotal}")

print("---------------------------------------------")
print(f"Total Amount = â‚¹{total}")

if discount > 0:
    print(f"Discount Applied = â‚¹{discount}")

print(f"Final Amount to Pay = â‚¹{final_amount}")
print("\nThank you for ordering! ðŸ˜Š")
```
